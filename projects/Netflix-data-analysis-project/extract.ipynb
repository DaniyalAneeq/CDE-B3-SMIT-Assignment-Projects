```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
```

```python
import sqlalchemy as sal
engine = sal.create_engine(
    'mssql+pyodbc://Dell-10th/master?driver=ODBC+Driver+17+for+SQL+Server',  fast_executemany=True
    )
conn= engine.connect()
```

```python
from sqlalchemy.types import NVARCHAR, Integer

df.to_sql(
    "netflix_data",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "show_id": NVARCHAR(10),
        "type": NVARCHAR(10),
        "title": NVARCHAR(200),
        "director": NVARCHAR(250),
        "cast": NVARCHAR(1000),
        "country": NVARCHAR(150),
        "date_added": NVARCHAR(20),
        "release_year": Integer(),
        "rating": NVARCHAR(10),
        "duration": NVARCHAR(10),
        "listed_in": NVARCHAR(100),
        "description": NVARCHAR(500),
    }
)
conn.close()
```
